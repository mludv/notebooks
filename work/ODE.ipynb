# Modeling Nonlinear Electrical Circuits

We consider a classical electromagnetical model in two variants:

$$
\begin{cases}
\dot{x} &= y \\
\dot{y} &= -x + y - y^3 
\end{cases}
\qquad
\text{and}
\qquad
\begin{cases}
\dot{x} &= y \\
\dot{y} &= -x - y - y^3 
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact
%matplotlib inline

# Definition of the two models
f1 = lambda y, t: (y[1], -y[0]+y[1]-y[1]**3)
f2 = lambda y, t: (y[1], -y[0]-y[1]-y[1]**3)
t = np.arange(0,10,0.01)

# Prepare figures
fig1 = plt.figure(figsize=(15,5))
ax1 = fig1.add_subplot(121)
ax1.set_title('Model 1')
ax1.axis((-1.5,1.5,-1.5,4.5))
ax2 = fig1.add_subplot(122)
ax2.set_title('Model 2')
ax2.axis((-1.5,1.5,-1.5,4.5))

# Integrate
for i in [-0.7, -0.5, 0.6, 1.5, 3, 4]:
    y0 = [-0.5, i]
    y = odeint(f1, y0, t)
    ax1.plot(y[:,0], y[:,1])
    y = odeint(f2, y0, t)
    ax2.plot(y[:,0], y[:,1])
    
plt.show()
```
